# Social Media Engagement Dashboard
Analyzing engagement metrics to recommend optimal posting strategies.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"D:\code\data\projects\full projects\Social Media Engagement Dashboard\data\mediarates.csv")
df.dropna(subset=['likes','comments','shares','engagement_rate'], inplace=True)
df['post_hour'] = df['post_hour'].astype(int)
```

## Engagement by Posting Hour

```python
hourly_engagement = df.groupby('post_hour')['engagement_rate'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=hourly_engagement.index, y=hourly_engagement.values, palette='viridis')
plt.title('Average Engagement Rate by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Engagement Rate')
plt.tight_layout()
plt.show()
```

## Engagement by Post Category

```python
category_engagement = df.groupby('post_category')['engagement_rate'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=category_engagement.values, y=category_engagement.index, palette='coolwarm')
plt.title('Average Engagement Rate by Post Category')
plt.xlabel('Engagement Rate')
plt.ylabel('Post Category')
plt.tight_layout()
plt.show()
```

## Engagement vs Followers

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='user_followers', y='engagement_rate', hue='user_verified', data=df, alpha=0.6)
plt.xscale('log')
plt.title('Engagement Rate vs Followers (Log Scale)')
plt.xlabel('Followers (log scale)')
plt.ylabel('Engagement Rate')
plt.legend(title='Verified')
plt.tight_layout()
plt.show()
```
